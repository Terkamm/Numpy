## Data loading

Load the dataset "Plateau PHC Board 1.csv" into a pandas DataFrame.

## Data exploration

### Subtask:
Explore the dataset to understand its structure, including column names, data types, and the presence of missing values. Check the distribution of pre-test and post-test scores.


**Reasoning**:
Examine the DataFrame's structure using .info(), check for missing values using .isnull().sum() and .describe(), and analyze the score distributions using .describe() and visualizations (histograms and box plots).



```python
# Examine DataFrame structure
print(df.info())

# Check for missing values
print(df.isnull().sum())
print(df.describe())

# Analyze score distributions
print(df[['PRE TEST SCORE', 'POST TEST SCORE']].describe())

import matplotlib.pyplot as plt

# Histograms
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['PRE TEST SCORE'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Pre-Test Scores')
plt.xlabel('Pre-Test Score')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['POST TEST SCORE'], bins=10, color='salmon', edgecolor='black')
plt.title('Distribution of Post-Test Scores')
plt.xlabel('Post-Test Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()


# Box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['PRE TEST SCORE'], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box Plot of Pre-Test Scores')

plt.subplot(1, 2, 2)
plt.boxplot(df['POST TEST SCORE'], vert=False, patch_artist=True, boxprops=dict(facecolor='salmon'))
plt.title('Box Plot of Post-Test Scores')
plt.tight_layout()
plt.show()
```

## Data cleaning

### Subtask:
Clean the pre-test and post-test score columns by handling any potential outliers.


**Reasoning**:
Examine the distributions of 'PRE TEST SCORE' and 'POST TEST SCORE' using box plots to identify potential outliers. Then, decide on a strategy for handling them (removal, capping, or transformation) and implement it.



```python
import matplotlib.pyplot as plt

# Box plots to visualize potential outliers
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['PRE TEST SCORE'], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box Plot of Pre-Test Scores')

plt.subplot(1, 2, 2)
plt.boxplot(df['POST TEST SCORE'], vert=False, patch_artist=True, boxprops=dict(facecolor='salmon'))
plt.title('Box Plot of Post-Test Scores')
plt.tight_layout()
plt.show()

# Based on the box plots, identify outliers.  For this example, let's assume any score below 10 is an outlier in both pre and post-test.
# Strategy: Capping - replace outliers with the lower bound (10)
df['PRE TEST SCORE'] = df['PRE TEST SCORE'].apply(lambda x: 10 if x < 10 else x)
df['POST TEST SCORE'] = df['POST TEST SCORE'].apply(lambda x: 10 if x < 10 else x)

# Verification: Display updated boxplots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['PRE TEST SCORE'], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box Plot of Pre-Test Scores (After Capping)')

plt.subplot(1, 2, 2)
plt.boxplot(df['POST TEST SCORE'], vert=False, patch_artist=True, boxprops=dict(facecolor='salmon'))
plt.title('Box Plot of Post-Test Scores (After Capping)')
plt.tight_layout()
plt.show()
```

## Data analysis

### Subtask:
Calculate descriptive statistics and perform a paired t-test.


**Reasoning**:
Calculate descriptive statistics for 'PRE TEST SCORE' and 'POST TEST SCORE', and perform a paired t-test.



```python
from scipy.stats import ttest_rel

# Calculate descriptive statistics
pre_test_stats = df['PRE TEST SCORE'].describe()
post_test_stats = df['POST TEST SCORE'].describe()

print("Descriptive Statistics for Pre-Test Scores:")
print(pre_test_stats)
print("\nDescriptive Statistics for Post-Test Scores:")
print(post_test_stats)

# Perform paired t-test
t_statistic, p_value = ttest_rel(df['PRE TEST SCORE'], df['POST TEST SCORE'])

print("\nPaired t-test results:")
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")
```

## Data visualization

### Subtask:
Visualize the pre-test and post-test scores using appropriate charts.


**Reasoning**:
Create the visualizations as requested in the instructions: scatter plot, histograms, box plot, and an optional bar chart.



```python
import matplotlib.pyplot as plt

# 1. Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['PRE TEST SCORE'], df['POST TEST SCORE'], color='purple', alpha=0.7)
plt.xlabel("Pre-Test Score")
plt.ylabel("Post-Test Score")
plt.title("Scatter Plot of Pre-Test vs. Post-Test Scores")
plt.grid(True)
plt.show()

# 2. Histograms
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['PRE TEST SCORE'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Pre-Test Scores')
plt.xlabel('Pre-Test Score')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['POST TEST SCORE'], bins=10, color='salmon', edgecolor='black')
plt.title('Distribution of Post-Test Scores')
plt.xlabel('Post-Test Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 3. Box plot of the difference
plt.figure(figsize=(6, 4))
plt.boxplot(df['Difference'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Distribution of Difference Between Pre and Post Test Scores')
plt.xlabel('Score Difference')
plt.show()

# 4. Bar chart of average difference (optional)
plt.figure(figsize=(6, 4))
plt.bar(x=['Average Difference'], height=df['Difference'].mean(), yerr=df['Difference'].std(), color='orange', capsize=5)
plt.title('Average Difference Between Pre and Post Test Scores')
plt.ylabel('Average Difference')
plt.show()
```
